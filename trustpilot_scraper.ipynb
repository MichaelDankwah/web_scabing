**Web Scraping Reviews from Trustpilot**

```python
# import necessary libraries
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Function to extract data from BeautifulSoup objects
def soup2list(src, list_, attr=None):
    if attr:
        for val in src:
            list_.append(val.get(attr))
    else:
        for val in src:
            list_.append(val.get_text(strip=True))

# Lists to store extracted data
users = []
userReviewNum = []
ratings = []
locations = []
dates = []
reviews = []

from_page = 1
to_page = 500
company = 'backmarket.co.uk'

# Loop through pages
for i in range(from_page, to_page + 1):
    print(f"Scraping page {i}...")

    # Send request to Trustpilot page
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    result = requests.get(f"https://www.trustpilot.com/review/{company}?page={i}", headers=headers)
    soup = BeautifulSoup(result.content, 'html.parser')

    # Temporary lists to check for equal length after each page
    tmp_users = []
    tmp_userReviewNum = []
    tmp_ratings = []
    tmp_locations = []
    tmp_dates = []
    tmp_reviews = []

    # Extract data using specified selectors
    soup2list(soup.find_all('span', {'class':'typography_heading-xxs__QKBS8 typography_appearance-default__AAY17'}), tmp_users)
    soup2list(soup.find_all('div', {'class':'typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_detailsIcon__Fo_ua'}), tmp_locations)
    soup2list(soup.find_all('span', {'class':'typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l'}), tmp_userReviewNum)
    soup2list(soup.find_all('div', {'class':'styles_reviewHeader__iU9Px'}), tmp_dates)
    soup2list(soup.find_all('div', {'class':'styles_reviewHeader__iU9Px'}), tmp_ratings, attr='data-service-review-rating')
    soup2list(soup.find_all('div', {'class':'styles_reviewContent__0Q2Tg'}), tmp_reviews)

    # Ensure consistent lengths
    max_len = max(len(tmp_users), len(tmp_userReviewNum), len(tmp_ratings), len(tmp_locations), len(tmp_dates), len(tmp_reviews))

    users.extend(tmp_users + [''] * (max_len - len(tmp_users)))
    userReviewNum.extend(tmp_userReviewNum + [''] * (max_len - len(tmp_userReviewNum)))
    ratings.extend(tmp_ratings + [''] * (max_len - len(tmp_ratings)))
    locations.extend(tmp_locations + [''] * (max_len - len(tmp_locations)))
    dates.extend(tmp_dates + [''] * (max_len - len(tmp_dates)))
    reviews.extend(tmp_reviews + [''] * (max_len - len(tmp_reviews)))

    # Sleep to avoid throttling
    time.sleep(2)

# Create a DataFrame from the extracted data
review_data = pd.DataFrame({
    'Username': users,
    'Total reviews': userReviewNum,
    'Location': locations,
    'Date': dates,
    'Content': reviews,
    'Rating': ratings
})

# Save DataFrame to a CSV file
review_data.to_csv(f'{company}_reviews.csv', index=False)
```
